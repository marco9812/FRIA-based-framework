## Competences e Knowledge

### CREAZIONE COLONNA COVERAGE PER COMPETENCES E KNOWLEDGE

```python
# funzione per calcolare la copertura
def calculate_coverage(cv_items, job_items):
    matched_items = set(cv_items).intersection(set(job_items))
    if len(job_items) == 0:  # Evita divisione per zero
        return 1 ##se non ci sono richieste, allora Ã¨ 1
    return len(matched_items) / len(job_items)
```

```python
# Applica la funzione per la copertura obbligatoria e opzionale (abbiamo due liste all'interno della colonna, una per le opzionali una per le obbligatorie)
df_cv_job['competences_coverage_required'] = df_cv_job.apply(lambda row: calculate_coverage(row['competences_cv'], row['competences_job'][0]), axis=1)
df_cv_job['competences_coverage_optional'] = df_cv_job.apply(lambda row: calculate_coverage(row['competences_cv'], row['competences_job'][1]), axis=1)

# eliminare le colonne competences_cv e competences_job 
df_cv_job = df_cv_job.drop(columns=['competences_cv', 'competences_job'])

# risultato
print(df_cv_job[['competences_coverage_required', 'competences_coverage_optional']])
```

## Languages

### Creazione colonna coverage per languages (tenendo conto del livello)

```python
# Definiamo la gerarchia dei livelli
livelli = ["NA", "A1", "A2", "B1", "B2", "C1", "C2"]
```

```python
def livello_superiore_o_uguale(livello_cv, livello_job):
    """Ritorna True se il livello del CV Ã¨ uguale o superiore a quello richiesto dal job."""
    if livello_cv not in livelli or livello_job not in livelli:
        return False
    return livelli.index(livello_cv) >= livelli.index(livello_job)
```

```python

def estrai_lingue_e_livelli(lista):
    """Dividi la stringa 'lingua,livello' in una coppia (lingua, livello)."""
    return dict(item.split(',') for item in lista)
```

```python
def calcola_copertura(row):
    # Estrai lingue e livelli dal CV
    lingue_cv = estrai_lingue_e_livelli(row['languages_cv'])
    
    # Estrai lingue obbligatorie e opzionali dal job
    lingue_obbligatorie_job = estrai_lingue_e_livelli(row['languages_job'][0])
    lingue_opzionali_job = estrai_lingue_e_livelli(row['languages_job'][1])
    
    # Calcola copertura per le lingue obbligatorie
    copertura_obbligatorie = sum(
        1 for lingua, livello_job in lingue_obbligatorie_job.items() 
        if lingua in lingue_cv and livello_superiore_o_uguale(lingue_cv[lingua], livello_job)
    ) / len(lingue_obbligatorie_job) if lingue_obbligatorie_job else 1.0
    
    # Calcola copertura per le lingue opzionali
    copertura_opzionali = sum(
        1 for lingua, livello_job in lingue_opzionali_job.items() 
        if lingua in lingue_cv and livello_superiore_o_uguale(lingue_cv[lingua], livello_job)
    ) / len(lingue_opzionali_job) if lingue_opzionali_job else 1.0
    
    return pd.Series([copertura_obbligatorie, copertura_opzionali])
```

```python
# Applicare la funzione al dataframe df_cv_job e creare le nuove colonne
df_cv_job[['copertura_lingue_obbligatorie', 'copertura_lingue_opzionali']] = df_cv_job.apply(calcola_copertura, axis=1)

# Visualizziamo le prime righe del dataframe aggiornato
df_cv_job.head()
```

## Prima di fare l'encoding delle colonne nationality e city, andiamo a creare delle correlazioni con le colonne religion e ethnicity per quanto riguarda le colonne _cv

```python
import random
```

```python
def get_cities_for_religion(religion, city_df):
    cities = city_df['city'].tolist()
    
    if religion == 'Christianity':
        return cities[:int(len(cities) * 0.25)]  # Supponiamo di prendere il 25% delle cittÃ 
    elif religion == 'Islam':
        return cities[int(len(cities) * 0.25):int(len(cities) * 0.5)]
    elif religion == 'Buddhism':
        return cities[int(len(cities) * 0.5):int(len(cities) * 0.75)]
    elif religion == 'Judaism':
        return cities[int(len(cities) * 0.75):]
    else:
        return cities  # Se la religione non corrisponde a nulla, usiamo tutte le cittÃ 
```

```python
# Funzione per creare la correlazione tra 'religion' e 'city'
def create_city_correlation(df, correlation_percent, city_df):
    cities = city_df['city'].tolist()  # Usa tutte le cittÃ  dal CSV
    for i in range(len(df)):
        religion = df.at[i, 'religion_cv']  # Prendi la religione corrente
        if random.random() < correlation_percent:
            # Prendi cittÃ  correlate in base alla religione
            religion_cities = get_cities_for_religion(religion, city_df)
            df.at[i, 'city_cv'] = random.choice(religion_cities)
        else:
            # Se non si applica la correlazione, scegli una cittÃ  casuale
            df.at[i, 'city_cv'] = random.choice(cities)
```

```python
nationalities = ['Spanish', 'Italian', 'French', 'English', 'Swedish', 'Irish', 'Russian', 'LowGerman', 'Romanian', 'Catalan', 'German']
```

```python
# Funzione per creare un mapping fittizio tra ethnicity e nazionalitÃ 
def get_nationalities_for_ethnicity(ethnicity):
    if ethnicity == 'White':
        return nationalities[:3]  # Ad esempio, White potrebbe essere associata a Spanish, Italian, French
    elif ethnicity == 'Black':
        return nationalities[3:6]  # Ad esempio, Black potrebbe essere associata a English, Swedish, Irish
    elif ethnicity == 'Asian':
        return nationalities[6:8]  # Ad esempio, Asian potrebbe essere associata a Russian, LowGerman
    elif ethnicity == 'Hispanic':
        return nationalities[8:]  # Ad esempio, Hispanic potrebbe essere associata a Romanian, Catalan, German
    else:
        return nationalities  # Se l'etnia non corrisponde a nulla, usiamo tutte le nazionalitÃ 
```

```python
# Funzione per creare la correlazione tra 'ethnicity' e 'nationality'
def create_nationality_correlation(df, correlation_percent):
    for i in range(len(df)):
        ethnicity = df.at[i, 'ethnicity_cv']  # Prendi l'etnia corrente
        if random.random() < correlation_percent:
            # Prendi nazionalitÃ  correlate in base all'etnia
            ethnicity_nationalities = get_nationalities_for_ethnicity(ethnicity)
            df.at[i, 'nationality_cv'] = random.choice(ethnicity_nationalities)
        else:
            # Se non si applica la correlazione, scegli una nazionalitÃ  casuale
            df.at[i, 'nationality_cv'] = random.choice(nationalities)
```

```python
# Applica la correlazione tra 'ethnicity' e 'nationality'
create_nationality_correlation(df_cv_job, 1)
```

```python
import numpy as np
import scipy.stats as stats
```

```python
def cramers_v(confusion_matrix):
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))
```

```python
# Calcolare V di Cramer
v_cramer_eth_nat = cramers_v(contingency_table_eth_nat.values)
v_cramer_religion_city = cramers_v(contingency_table_religion_city.values)
```

## Name e Surname

Utilizziamo Word2Vec (libreria gensim) per rappresentare name e surname come vettori continui e catturarne la semantica

```python
from gensim.models import Word2Vec
```

```python
# Unire nome e cognome
df_cv_job["Full_Name"] = df_cv_job["Generated_Name_cv"] + " " + df_cv_job["Generated_Surname_cv"]

# Preprocessing: tokenizzare i nomi completi in frasi
sentences = [full_name.split() for full_name in df_cv_job["Full_Name"]]

# Addestrare il modello Word2Vec
w2v_model = Word2Vec(
    sentences=sentences, 
    vector_size=50,  # Dimensione dei vettori
    window=5,        # Finestra di contesto
    min_count=1,     # Minimo numero di occorrenze per una parola
    sg=1             # Skip-gram (sg=1) o CBOW (sg=0)
)
```

```python

# Funzione per ottenere il vettore di una parola
def get_word_vector(word):
    return w2v_model.wv[word] if word in w2v_model.wv else [0] * 50
```

```python
# Funzione per ottenere un vettore combinato per il nome completo
def get_full_name_vector(full_name):
    words = full_name.split()
    vectors = [get_word_vector(word) for word in words]
    # Somma o media dei vettori delle parole
    return [sum(x) / len(x) for x in zip(*vectors)] if vectors else [0] * 50
```

```python
# Applicare Word2Vec alla colonna Full_Name
df_cv_job["Full_Name_Vector"] = df_cv_job["Full_Name"].apply(get_full_name_vector)
```

```python
# Espandere i vettori in colonne
vector_columns = pd.DataFrame(df_cv_job["Full_Name_Vector"].tolist(), columns=[f"Vec_{i}" for i in range(1, 51)])
df_cv_job = pd.concat([df_cv_job.drop(columns=["Full_Name_Vector", "Full_Name", "Generated_Name_cv","Generated_Surname_cv"]), vector_columns], axis=1)
```
